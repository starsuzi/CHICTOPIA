```python
import codecs
```

```python
s_file = codecs.open('./sample.txt', 'rb', 'utf-8')
#filedata = filedata.replace("',", ("',"+'\n'))
#filedata = filedata.replace("],", ("'],"+'\n'))
#with open('./sample.txt', 'w') as file :
#    file.write()
```

```python
s_filedata = s_file.read()   
s_filedata = s_filedata.replace('}]', ('}]'+'\n'))
s_filedata = s_filedata.replace("{", '')
s_filedata = s_filedata.replace("}", '')
s_filedata = s_filedata.replace("'", '')
s_filedata = s_filedata.replace("post_id", '\n'+'[post_id')
s_filedata = s_filedata.replace("title", '\n'+'title')
s_filedata = s_filedata.replace("photographer", '\n'+'photographer')
s_filedata = s_filedata.replace("lst_url", '\n'+'url')
s_filedata = s_filedata.replace(" lst_size", '\n'+'size')
s_filedata = s_filedata.replace("lst_tag", '\n'+'tag')
s_filedata = s_filedata.replace("lst_item", '\n'+'item]')
s_filedata = s_filedata.replace("[", '')
s_filedata = s_filedata.replace("]", '')

s_filedata = s_filedata.replace("http", '\n'+'http')
s_filedata = s_filedata.replace("jpg,", 'jpg')
```

```python
print(s_filedata)
```

```python
f_file = codecs.open('./sample.txt', 'wb', 'utf-8')
f_file.write(s_filedata)
```

```python
lst_filedata_ex = []
file_ex = codecs.open('./ex.txt', 'rb', 'utf-8')
filedata_ex = file_ex.readlines()   
#filedata = filedata.replace('}]', ('}]'+'\n'))
for line in filedata_ex:
    if 'http' in line:
        print(line)
        lst_filedata_ex.append(line)
```

```python
lst_filedata_ex
```

```python
import re
lst_num = []
for url in lst_filedata_ex:
    #print(url[38:43])
    url = str(url)
    print(re.findall('\d+', url))
    lst_num.append(re.findall('\d+', url))
```

```python
lst_num_final = []
for num in lst_num:
    print(num[0])
    lst_num_final.append(int(num[0]))
```
